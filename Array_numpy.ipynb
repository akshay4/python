```python
l1 = [1,2,3,4]
l2 = [5,6,7,8]
l1 + l2
```

```python
def add_1d(arr1,arr2):
    l = len(arr1)
    r = []
    for index in range(l):
        x = l1[index]+l2[index]
        r.append(x)
    return r
```

```python
r = add_1d(l1,l2)
r
```
